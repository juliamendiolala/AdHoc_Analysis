**SPLIT TRANSACTION_ITEMS COLUMN ONE ITEM PER ROW**

```python
# helper funaction to split transaction items to colon
def pir(df, col):
    colc = df[col].str.split(';')
    clst = colc.values.tolist()
    lens = [len(l) for l in clst]
    cdf = pd.DataFrame({col: np.concatenate(clst)}, df.index.repeat(lens))
    
    return df.drop(col, 1).join(cdf).reset_index(drop=True)


splitted = pir(df,'transaction_items')
# add column item_count
splitted['item_count'] = splitted['transaction_items'].apply(lambda x: int(x.split(',')[-1].replace('(x','').replace(')','')))

# apply function to column transaction_items convert to list to use the explode function
splitted['transaction_items'] = splitted['transaction_items'].apply(lambda x:x.split(',')[:-1])
splitted = splitted.explode('transaction_items')
# drop transaction value for we are dealing items 1 by 1
item_sales = splitted.drop(columns='transaction_value')
item_sales
```

**CONVERT TRANSACTION_DATE COLUMN TO MONTH (IN WORDS)**

```python
month = {'01':'January','02':'February','03':'March','04':'April','05':'May','06':'June'}
item_sales['transaction_date'] = splitted['transaction_date'].apply(lambda x: month[x.split('/')[1]])
item_sales
```

**GET TOTAL SOLD PER ITEM PER MONTH (COLUMN) AND TOTAL SOLD ITEMS PER MONTH (ROW)**

```python
item_sold = pd.pivot_table(item_sales,columns='transaction_date',values='item_count',index='transaction_items',aggfunc='sum')
item_sold = item_sold.reindex(columns=month.values())
total_sale = [item_sold['January'].sum(),
                          item_sold['February'].sum(),
                          item_sold['March'].sum(),
                          item_sold['April'].sum(),
                          item_sold['May'].sum(),
                          item_sold['June'].sum()]
item_sold.loc['Total Items Sold'] = total_sale
item_sold['Total Sold'] = item_sold.sum(axis=1)
item_sold
```

**GET TOTAL SALES PER MONTH**

```python
#Get total sales per month by adding the transaction_value per month

summary_sale = df.drop(columns=['transaction_items','username','sex'])
summary_sale['transaction_date'] = summary_sale['transaction_date'].apply(lambda x: month[x.split('/')[1]])
summary_sale = pd.pivot_table(summary_sale,columns='transaction_date',values='transaction_value',aggfunc='sum')
columns = ["January", "February", "March", "April", "May", "June"]
summary_sale = summary_sale[columns]
summary_sale
```

**GET TOTAL COUNT PER AGE CATEGORY**

```python
def age(born):
  born = datetime.strptime(born,"%Y/%m/%d").date()
  today = datetime.today()
  return today.year-born.year - ((today.month,today.day)<(born.month,born.day))

def cat_age(age):
  base = {'Children':range(0,14),'Teen':range(14,20), 'Youth':range(20,25), 'Adults':range(25,64), 'Senior':range(64,120)}
  for k,v in base.items():
    if int(age) in v:
      return k

df2 = pd.read_json("transaction-data-adhoc-analysis.json")

df2['Age'] = df2.birthdate.apply(age)
agedf = df2.drop(columns=['transaction_items',	'transaction_value',	'transaction_date','address','mail'])
agedf = agedf.drop_duplicates('username')
agedf['Age Category'] = agedf.Age.apply(cat_age)

grouped = agedf.groupby('Age Category').count()

ind = ['Children', 'Teen', 'Youth', 'Adults', 'Senior']
grouped = grouped.reindex(ind)

cols = [3,2,1,0]
grouped.drop(grouped.columns[cols],inplace=True,axis=1)

grouped.rename(columns = {'Age':'Count'}, inplace = True)

grouped
```

**GET NUMBER OF REPEATERS, INACTIVE, AND ENGAGED CUSTOMERS**

```python
# create new dataframe with columns are months
lolaDF = pd.DataFrame(columns = list(month.values()))

#get the unique customer in evry month

customer = item_sales.drop_duplicates(subset=['username','transaction_date'], keep='first')
customer.rename(columns = {'transaction_date':'month'}, inplace = True)
# extract customers in every month
jan_cus = customer.loc[customer['month'] == 'January'].username 
feb_cus = customer.loc[customer['month'] == 'February'].username
mar_cus = customer.loc[customer['month'] == 'March'].username
apr_cus = customer.loc[customer['month'] == 'April'].username
may_cus = customer.loc[customer['month'] == 'May'].username
june_cus = customer.loc[customer['month'] == 'June'].username

# lets calculate repeaters
feb_rep = feb_cus.isin(jan_cus)
mar_rep = mar_cus.isin(feb_cus)
apr_rep = apr_cus.isin(mar_cus)
may_rep = may_cus.isin(apr_cus)
june_rep = june_cus.isin(may_cus)
repeaters = [0,len([x for x in feb_rep.values if x == True]),
             len([x for x in mar_rep.values if x == True]),
             len([x for x in apr_rep.values if x == True]),
             len([x for x in may_rep.values if x == True]),
             len([x for x in june_rep.values if x == True])]

lolaDF.loc['Repeaters'] = repeaters
# lets calculate Inactive

feb_inac = len([x for x in (jan_cus.isin(feb_cus)).values if x == False])
mar_inac = len([x for x in (feb_cus.isin(mar_cus)).values if x == False])
apr_inac = len([x for x in (mar_cus.isin(apr_cus)).values if x == False])
may_inac = len([x for x in (apr_cus.isin(may_cus)).values if x == False])
june_inac = len([x for x in (may_cus.isin(june_cus)).values if x == False])
inactive = [0,feb_inac,mar_inac,apr_inac,may_inac,june_inac]
lolaDF.loc['Inactive'] = inactive

# Calculate Customer Consistently purchased
jan_con = jan_cus
feb_con = jan_con.loc[jan_con.map(lambda x: x in feb_cus.values)]
mar_con = feb_con.loc[feb_con.map(lambda x: x in mar_cus.values)]
apr_con = mar_con.loc[mar_con.map(lambda x: x in apr_cus.values)]
may_con = apr_con.loc[apr_con.map(lambda x: x in may_cus.values)]
june_con = may_con.loc[may_con.map(lambda x: x in june_cus.values)]

engaged = [jan_cus.count(),
              feb_con.count(),
              mar_con.count(),
              apr_con.count(),
              may_con.count(),
              june_con.count(),
              ]
              

lolaDF.loc['Engaged'] = engaged

lolaDF
```

**CREATE NUMBER OF MALES AND FEMALES PER MONTH GRAPH (GROUPED BAR GRAPH)**

```python
#count number of males and females per month

#set width of the bar
bar_width = 0.4

#list all months
months = ["January", "February", "March", "April", "May", "June"]

#list the values
male = [6892,6966,7061,6928,7014,7002]
female = [6796,7006,6841,6851,6850,6829]

#define the length of the months
index = np.arange(len(months))

#declare the bar graph
fig, ax = plt.subplots()
bar_male = ax.bar(index - bar_width/2, male, bar_width, label = "Male")
bar_female = ax.bar(index + bar_width/2, female, bar_width, label = "Female", color = "pink")

#change the label of the x-axis to months
ax.set_xticks(index)
ax.set_xticklabels(months)

#change labels and titles
ax.set_xlabel("Months")
ax.set_ylabel("Count")
ax.set_title("Male vs Female Customers")

#create function for inserting value labels on the bar graph
def value_label(bars):
    for bar in bars:
        bar_height = bar.get_height()
        ax.annotate('{0:.0f}'.format(bar.get_height()),
            xy=(bar.get_x() + bar.get_width()/2, bar_height),
            xytext = (0,3),
            textcoords = 'offset points',
            ha = 'center',
            va = 'bottom'
        )

#use the function to use it on bar_male and bar_female
value_label(bar_male)
value_label(bar_female)

#set the limit of y-axis
plt.ylim([6400, 7200])

#display the graph
plt.legend()
plt.show()
```
